```python
correlation = df["Age"].corr(df["Premium"])
print("Correlation Coefficient:", correlation)
```

```python
X=df.drop('Premium',axis=1)
y=df['Age']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
LR.fit(X_train,y_train)
```

```python
y_pred=LR.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
M=mean_squared_error(y_test,y_pred)
M
```

```python
Absolute=mean_absolute_error(y_test,y_pred)
Absolute
```

```python
from math import sqrt
rms = sqrt(mean_squared_error(y_test, y_pred))
rms
```

```python
R2_score=r2_score(y_test,y_pred)
R2_score
```

```python
plt.scatter(X_train, y_train, alpha=0.7, label='Training data', color='red')
plt.scatter(X_test, y_test, alpha=0.7, label='Testing Data', color='green')
plt.xlabel('Age')
plt.ylabel('Premium')
plt.legend()
```
