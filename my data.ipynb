```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
df_grades = pd.read_csv("superstore.csv")
df_grades['Country'].unique
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x='Region', y='Quantity', data= df_grades)
plt.title('Quantity purchased per Region')
plt.show()
```

```python
mean_sales_by_category = df_grades.groupby('Discount')['Sales'].mean()

# Plot a pie chart
plt.figure(figsize=(10, 80))
plt.pie(mean_sales_by_category, labels=mean_sales_by_category.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Mean Sales by Discount')
```

```python
mean_sales_by_customer = df_grades.groupby('Customer.Name')['Sales'].mean()


# Create a KDE plot for the mean number of sales
plt.figure(figsize=(10, 6))
sns.kdeplot(mean_sales_by_customer)

# Add labels and a title
plt.xlabel('Mean Number of Sales')
plt.ylabel('Density')
plt.title('Kernel Density Estimate of Mean Sales by Customer')
```

```python
product_counts = df_grades['Product.Name'].value_counts()

# Extract the product with the highest count
top_product = product_counts.idxmax()
highest_count = product_counts.max()

print(f"The products with names {top_product} was bought the most, with a count of {highest_count} as assigned to each.")
plt.figure(figsize=(20, 6))
sns.barplot(x=product_counts.index[:6], y=product_counts.values[:6], palette="viridis")

plt.xlabel('Product ID')
plt.ylabel('Number of Purchases')
plt.title('Top 10 Products Bought the Most')
```

```python
total_sales_by_product = df_grades.groupby('Product.Name')['Sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Product.Name', y='Sales', data=total_sales_by_product)
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.title('Correlation between Products and Sales')
plt.show()
```

```python
df_grades['Order.Date'] = pd.to_datetime(df_grades['Order.Date'])
order_counts_by_date = df_grades['Order.Date'].value_counts()
date_with_highest_orders = order_counts_by_date.idxmax()
highest_orders_count = order_counts_by_date.max()
print(f"The date with the highest number of orders is {date_with_highest_orders} with {highest_orders_count} orders.")
plt.figure(figsize=(12, 6))
sns.barplot(x=order_counts_by_date.index[:10], y=order_counts_by_date.values[:10], palette="viridis")
plt.xlabel('Order Date')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Date')
plt.xticks(rotation=45, ha='right')
plt.show()
```
